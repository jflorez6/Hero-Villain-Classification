```python
import requests
import shutil
from marvel import Marvel
```

```python
filename = 'apikey'
```

```python
def get_file_contents(filename):
    try:
        with open(filename, 'r') as f:
            api_keys = f.readlines()
            return api_keys[0].replace('\n',''), api_keys[1]
    except FileNotFoundError:
        print("File not found!")
```

```python
public_key, private_key = get_file_contents(filename)
```

```python
m = Marvel(public_key, private_key)
```

```python
characters = m.characters.all(limit=100)
```

```python
total_char = characters['data']['total'] # total number of characters in Marvel Universe
print('Number of Characters in Marvel Universe:', total_char)
```

```python
# Get 100 images

image_variant = 'standard_large'

for i in range(total_char // 100 + 1):
    if i < total_char // 100: 
        n_iter = 100
    elif i == total_char // 100: 
        n_iter = total_char - (i * 100)
    characters = m.characters.all(limit=n_iter,offset=i*100)
    for j in range(n_iter):
        image = characters['data']['results'][j]['thumbnail']
        name = characters['data']['results'][j]['name']
        name = name.replace(" ", "_")
        name = name.replace("/", "_")
        filename = name + '.' + image['extension']
        image_url = image['path'] + '/' + image_variant + '.' + image['extension']
        r = requests.get(image_url, stream=True)
        if r.status_code == 200:
        # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
            r.raw.decode_content = True
            with open('data/' + filename, 'wb') as f:
                shutil.copyfileobj(r.raw, f)
            print('Image successfully downloaded: ', filename)
        else:
            print('Image could not be retrieved')
            
```
